```python
from matplotlib import image
from statistics import mean
from matplotlib import pyplot as plt
```

```python

img_name = "putin" # just image name
img_exp = ".jpg"
img = image.imread(img_name + img_exp)
```

```python
plt.imshow(img)
plt.show()
```

```python
def get_average_for_colour(test, x, y, z, d, num, h):
    mas = []
    for i in range(0, h):
        mas.append(mean(test[x:y,z:d,num][i]))
    av = mean(mas)
    return av

def get_average(test, x, y, z, d, h):
    return list(map(lambda i: get_average_for_colour(test, x, y, z, d, i, h), range(0, 3)))
```

```python
masked_img = img.copy()

plt.imshow(img)
plt.show()

delta = 10
y_min = 10
x_min = 135
y_max = 300
x_max = 335


while y_min < y_max:
    while x_min < x_max:
        mas = get_average(masked_img, y_min, y_min + delta, x_min, x_min + delta, delta)
        for i in range(0, 3):
            masked_img[y_min:(y_min + delta),x_min:(x_min + delta),i] = mas[i]
        x_min = x_min + delta
    y_min = y_min + delta
    x_min = 135

plt.imshow(masked_img)
plt.show()
image.imsave(img_name + '_masked' + img_exp, masked_img)
```

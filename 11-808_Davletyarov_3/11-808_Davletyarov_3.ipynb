```python
from matplotlib.image import imread
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from matplotlib import image
```

```python
img_name = "putin" # just image name
img_exp = ".jpg"
img = image.imread(img_name + img_exp)
```

```python
plt.imshow(img)
plt.show()
```

```python
def get_new_dataset(data, alpha):
    x = len(data[:,])
    y = len(data[0])

    X = np.empty((x, y, 5))
    X[:,:,0:3] = data[:,:,0:3]
    for i in range(0, x):
        for j in range(0, y):
            X[i, j, 3] = i * alpha
            X[i, j, 4] = j * alpha
    return X
```

```python
def clusterize(alpha, clusters):
  x = get_new_dataset(img, alpha)
  x_t = x.reshape((-1,img.shape[2]))

  kmeans = KMeans(n_clusters=clusters).fit(x_t)
  labels = kmeans.labels_
  centroids = kmeans.cluster_centers_

  newdata = centroids[labels]
  new_img = newdata.astype(int).reshape(x.shape)
  plt.imshow(new_img[:,:,0:3])
  plt.show()
  print(f'Clusterize with alpha = {alpha}, clusters = {clusters}')
```

```python
%%time
clusterize(0.5, 20)
```

```python
%%time
clusterize(0.9, 20)
```

```python
%%time
clusterize(0.1, 20)
```
